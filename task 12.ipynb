```python
import pandas as pd
import pickle
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

if 'test_y' not in globals():
	data = pd.read_csv('cleaned_data.csv')
	x = data.drop('selling_price', axis=1)
	y = data['selling_price']
	_, test_x, _, test_y = train_test_split(x, y, test_size=0.2, random_state=85)

if 'model_svc_pred' not in globals():
	if 'model_svc' not in globals():
		model_svc = pickle.load(open('model_svc.pkl', 'rb'))
	model_svc_pred = model_svc.predict(test_x)

model_svc_acc = accuracy_score(test_y, model_svc_pred)
print("-- Model Accuracy Score: ", end='')
print(round(model_svc_acc, 2) * 100)
```

```python
from typing import cast

import pandas as pd
import numpy as np
import pickle
import joblib

# Load Model & Encoder
model = pickle.load(open('model_svc.pkl', 'rb'))
encoders = joblib.load('encoders.pkl')

# Encoding helper
def encode_cols(le, value):
    # If the value is already a numeric encoded label and the encoder classes are non-numeric,
    # assume the value is already encoded.
    if isinstance(value, (int, np.integer, float, np.floating)) and le.classes_.dtype.kind not in 'iuf':
        return int(value)
    if value in le.classes_:
        return le.transform([value])[0]
    raise ValueError(f"Value {value!r} is not in encoder classes for this column.")

# Example input values for notebook execution
user_input = {
    "name": 0,  
    "year": 2016.0,
    "selling_price": 350000,
    "km_driven": 150000,
    "fuel": 2,  # Assuming 'fuel' is encoded as 2 for a specific fuel type
    "seller_type": 2,
    "transmission": 0.0,
    "owner": 3,
    "seats": 5.0,
    "max_power (in bph)": 1000,
    "Mileage Unit": 1,  # Assuming 'Mileage Unit' is encoded as 1 for a specific unit
    "Mileage": 20.0,
    "Engine (CC)": 1700
}
for col in ['Mileage Unit']:
    user_input[col] = encode_cols(encoders[col], user_input[col])

features = [
    "name",
    "year",
    "km_driven",
    "fuel",
    "seller_type",
    "transmission",
    "owner",
    "seats",
    "max_power (in bph)",
    "Mileage Unit",
    "Mileage",
    "Engine (CC)"
]

x = np.array([[user_input[cols] for cols in features]])
x = pd.DataFrame([user_input], columns=features)
predicted_price = model.predict(x)[0]
print(f"Predicted Selling Price: {predicted_price}")    
```

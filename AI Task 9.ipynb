```python
import math
def minimax(curDepth, nodeIndex, isMaxTurn, scores, targetDepth):
    if curDepth == targetDepth:
        return scores[nodeIndex]
    if isMaxTurn:
        left_child = minimax(curDepth + 1, nodeIndex * 2, False, scores, targetDepth)
        right_child = minimax(curDepth + 1, nodeIndex * 2 + 1, False, scores, targetDepth)
        return max(left_child, right_child)
    else:
        left_child = minimax(curDepth + 1, nodeIndex * 2, True, scores, targetDepth)
        right_child = minimax(curDepth + 1, nodeIndex * 2 + 1, True, scores, targetDepth)
        return min(left_child, right_child)
scores = [3, 5, 2, 9, 3, 5, 2, 9]   
treeDepth = int(math.log(len(scores), 2))  
print("The optimal value is:", minimax(0, 0, True, scores, treeDepth))
```

The Minimax algorithm is like 2 players taking turns: one always tries to get the **highest score** (Maximizer), while the other tries to get the **lowest score** (Minimizer). At each step, the algorithm looks ahead at possible outcomes and picks the best move for that player. In the end, it gives the **optimal guaranteed result** assuming both play smart.

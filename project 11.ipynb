```python
from sklearn.preprocessing import LabelEncoder
encoders = {}

def encodeCols(cols):
    for i in cols:
        value = pd.DataFrame({i:data[i].unique()})
        LE = LabelEncoder()
        LE.fit(value)
        data[i] = LE.transform(data[i])
        encoders[i] = LE

columns = ['name', 'year', 'fuel', 'seller_type', 'transmission', 'owner', 'Mileage Unit']
encodeCols(columns)
```

```python
data['Engine (CC)'] = data['Engine (CC)'].fillna(data['Engine (CC)'].mean())
```

```python
import joblib
# Save encoders
joblib.dump(encoders, "encoders.pkl")
```

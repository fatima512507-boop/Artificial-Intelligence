```python
from sklearn.preprocessing import LabelEncoder
encoders = {}

def encodeCols(cols):
    for i in cols:
        value = pd.DataFrame({i:data[i].unique()})
        LE = LabelEncoder()
        LE.fit(value)
        data[i] = LE.transform(data[i])
        encoders[i] = LE

columns = ['study_hours_per_week', 'sleep_hours_per_day', 'attendance_percentage',
       'assignments_completed', 'final_grade', 'participation_level_Low',
       'participation_level_Medium', 'internet_access_Yes',
       'parental_education_High School',
       'parental_education_PhD', 'extracurricular_Yes', 'part_time_job_Yes']
encodeCols(columns)
```

```python
data['attendance_percentage'] = data['attendance_percentage'].fillna(data['attendance_percentage'].mean())
```

```python
import joblib
# Save encoders
joblib.dump(encoders, "encoders.pkl")
```
